### Stematic sampling.
#### This samples the population at regular intervals.

````python
sample_size = 100
pop_size = len(df_coffee)
pop_size
````

````python
interval = pop_size // sample_size
interval
````

#### Thus, to get a **systematic sample of 5 coffees, we will select every 267 coffee** in the dataset.

````python
df_coffee.iloc[::interval]
````

#### Systematic sampling has a problem: **if the data has been sorted, or there is some sort of pattern or meaning behind the row order, then the resulting sample may not be representative of the whole population**.

````python
df_coffee_index_resets = df_coffee.reset_index()
df_coffee_index_resets.plot(x='index', y='aftertaste', kind='scatter')
plt.show()
````

#### **Earlier rows generally have higher aftertaste scores than later rows.** This introduces bias into the statistics that we calculate. 

#### To ensure that systematic sampling is safe, we can randomize the row order before sampling. dot-sample has an argument named frac that lets us specify the proportion of the dataset to return in the sample, rather than the absolute number of rows that n specifies. 

````python
shuffled = df_coffee.sample(frac=1) # shuffling the whole dataset
````

````python
shuffled.head()
````

````python
shuffled = shuffled.reset_index(drop=True)
````

````python
shuffled.head()
````

````python
shuffled = shuffled.reset_index()
````

````python
shuffled.head()
````

````python
shuffled.plot(x='index', y='aftertaste', kind='scatter')
plt.show()
````

````python
top_country = ['Mexico', 'Colombia', 'Guatemala', 'Brazil', 'Taiwan', 'United States (Hawaii)']
````

````python
top_country_subset = df_coffee['country_of_origin'].isin(top_country)
````

````python
df_coffee_top_country = df_coffee[top_country_subset]
````

````python
df_coffee_top_country.head()
````

### Lets Say We want a Sample that will represent Every Country

````python
coffee_sample_of_top_country = df_coffee_top_country.sample(frac=0.1, random_state=2021)
````

````python
coffee_sample_of_top_country['country_of_origin'].value_counts(normalize=True) * 100
````

````python
df_coffee['country_of_origin'].value_counts(normalize=True) * 100
````

````python
Proportional_stratified_sample = df_coffee.groupby('country_of_origin').sample(frac=0.1, random_state=2021)
````

### The Sample looks Closer to Original Population

````python
Proportional_stratified_sample['country_of_origin'].value_counts(normalize=True) * 100
````

### Now We Want a Sample where Every Country will be Same/Equal Proportion
#### sample() ```n``` parameter for select individual sample from each country

````python
equal_proportional_stratified_sample = df_coffee_top_country.groupby("country_of_origin").sample(n=15, random_state=2021)
equal_proportional_stratified_sample['country_of_origin'].count_values()
````

````python
equal_proportional_stratified_sample['country_of_origin'].count_values(normalize=True)
````

## Weighted Random Sampling
#### By setting custom Weight to a specific coulmn. 

````python
df_coffee_weight = df_coffee
conditon = df_coffee_weight['country_of_origin'] == 'Taiwan'
````

### Now I am Setting Weight by Condition and chosing 1 out of 2

````python
df_coffee_weight['weight'] = np.where(conditon, 2, 1)
````

````python
df_coffee_weight.head()
````

````python
df_coffee_weight = df_coffee_weight.sample(frac=0.1, weights='weight')
df_coffee_weight['country_of_origin'].value_counts(normalize=True)
````
