### What is Sampling?

#### Sampling is a process in statistical analysis where researchers take a predetermined number of observations from a larger population. The method of sampling depends on the type of analysis being performed.

```python
df_songs = pd.read_feather('data/spotify_2000_2020.feather')
df_songs.head()
```

```python
samled_df = df_songs.sample(n=1000)
samled_df['duration_minutes'].mean()
```

### Beta Distribution
#### In probability theory and statistics, the beta distribution is a **family of continuous probability distributions** defined on the interval [0, 1] in terms of two positive parameters, denoted by alpha and beta, that appear as exponents of the random variable and control the shape of the distribution.

```python
np.random.seed(43)

# Generate random numbers from a Normal(5, 2)
normals = np.random.normal(loc=5, scale=2, size=5000)

# Plot a histogram of normal values, binwidth 0.5
plt.hist(normals, bins=np.arange(-2, 13.5, 0.5))
plt.show()
```

### Stematic sampling.
#### This samples the population at regular intervals.

```python
sample_size = 100
pop_size = len(df_coffee)
pop_size
```

```python
interval = pop_size // sample_size
interval
```

#### Thus, to get a **systematic sample of 5 coffees, we will select every 267 coffee** in the dataset.

```python
df_coffee.iloc[::interval]
```

#### Systematic sampling has a problem: **if the data has been sorted, or there is some sort of pattern or meaning behind the row order, then the resulting sample may not be representative of the whole population**.

```python
df_coffee_index_resets = df_coffee.reset_index()
df_coffee_index_resets.plot(x='index', y='aftertaste', kind='scatter')
plt.show()
```

#### **Earlier rows generally have higher aftertaste scores than later rows.** This introduces bias into the statistics that we calculate. 

#### To ensure that systematic sampling is safe, we can randomize the row order before sampling. dot-sample has an argument named frac that lets us specify the proportion of the dataset to return in the sample, rather than the absolute number of rows that n specifies. 

```python
shuffled = df_coffee.sample(frac=1) # shuffling the whole dataset
```
