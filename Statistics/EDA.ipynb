## Coefficient **Correlation**

```python
# libraries
import scipy.stats as st
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the data
data = sns.load_dataset('taxis')
display(data.head())
```

### **Correlation**(using Numpy)

```python
print(np.corrcoef(data.distance, data.total))
```

```python
sns.scatterplot(x=data.distance, y=data.total)
plt.show()
```

### **Using scipy.stats**

```python
print(st.pearsonr(data.distance, data.total))
print('----------------------------x------------------------', '\n')
print(st.spearmanr(data.distance, data.total))
```

## Quick Checking of **Correlation** using **Seaborn**

```python
sns.pairplot(data)
plt.show()
```

```python
plt.figure(figsize=[12, 8])
sns.heatmap(data.corr())
plt.show()
```

## Quick Checking of **Correlation** using **Pandas**

```python
display(data.corr())
```
