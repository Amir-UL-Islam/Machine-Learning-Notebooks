````python
string = 'Our Deeds are the Reason of this #earthquake May ALLAH Forgive us all'
strings = [''.join(i) for i in df['Content'][:10]]
````

````python
# Libries for Sentiment Analysis
from transformers import pipeline
````

#### For Details about ```pipeline``` [huggingface](https://huggingface.co/docs/transformers/main_classes/pipelines)

````python
# Creating an Instance
classifier = pipeline('sentiment-analysis')
````

````python
result = classifier(string)
result
````

### Sentiment Analysis for list of Statement

````python
result = classifier(strings)
result
````

````python
# Libries for Sentiment Analysis
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from scipy.special import softmax
````

````python
regex_tokenized = []
final_tokenized_strings = []

# Tokenizer
for words in strings:
    for word in words.split(' '):
        word = '@user' if word.startswith('@') and len(word) > 1 else word
        word = 'http' if word.startswith('http') else word
        word = 'hastag' if word.startswith('#') else word
        regex_tokenized.append(word)
    string = ''.join(regex_tokenized)
    final_tokenized_strings.append(string)
final_tokenized_strings[1]
````

#### I am going use [Twitter-roBERTa-base for Sentiment Analysis](https://huggingface.co/cardiffnlp/twitter-roberta-base-sentiment). You can use any models from [Here](https://huggingface.co/models). 

````python
# Model Name
roberta = "cardiffnlp/twitter-roberta-base-sentiment"

#
model = AutoModelForSequenceClassification.from_pretrained(roberta)

#
tokenizer = AutoTokenizer.from_pretrained(roberta)
````

````python
tokenizer.model_max_length
````

````python
# labels
labels = ['Negative', 'Neutral', 'Positive']
````

### Sentimental Analysis

````python
# Encoding Tweets
encoded_tweets = []
outputs = []

# 
for tweets in final_tokenized_strings:
    encoded_tweets.append(tokenizer(tweets, return_tensors='pt'))

# Collecting Outputs
for index in range(len(encoded_tweets)):
    outputs.append(model(**encoded_tweets[index]))
````

### Scores


````python
scores = []
for o in outputs:
    scores.append(softmax(o[0][0].detach().numpy()))

len(scores)
for i in range(len(scores)):
    for j in range(len(scores[i])):
        l = labels[j]
        s = scores[i][j]
        print(l, s)
    print('\n')
````
